```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
# Project : Simulating Coin Flips with Binomial Distribution
```

```python
# Parameters
n = 10 # Number of coin flips
p = 0.5 # Probablility of heads
```

```python
# Generate possible number of heads
k = np.arange(0, n+1)
```

```python
# Calculate probabilites using binomial distribution
probabilities = binom.pmf(k, n, p)
```

```python
# Plot the probability mass function(PMF)
plt.bar(k, probabilities)
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

```python
# Simulate coin flips
num_simulations = 10000
flips = np. random.binomial(n, p, size=num_simulations)
```

```python
# Calculate and plot the histogram of the simulated results
plt.hist(flips, bins=n+1, density=True)
plt.xlabel("Number of Heads")
plt.ylabel("Frequency")
plt.title(f"Simulated Coin Flips (n={n}, p={p})")
plt.show
```

```python
# Calculate mean and standard deviation
mean  = n * p
std_dev = np.sqrt(n * p * (1 - p))

print (f"Mean: {mean}")
print (f"standard Deviation: {std_dev}")
```
