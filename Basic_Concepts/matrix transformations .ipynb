```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 2D vector
vector = np.array([1, 0])

# Define the rotation matrix
def rotation_matrix(theta):
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])

# Rotate the vector
theta = np.pi / 4  # 45 degrees
rot_matrix = rotation_matrix(theta)
rotated_vector = rot_matrix @ vector

# Plot the original and rotated vectors
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated Vector')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid()
plt.legend()
plt.title('2D Rotation Transformation')
plt.show()
```

```python
# Define the function and its gradient
def f(x):
    return x**2 + 4*x + 4

def gradient(x):
    return 2*x + 4

# Gradient Descent parameters
learning_rate = 0.1
x = 10  # Initial guess
iterations = 50

# Perform Gradient Descent
history = []
for _ in range(iterations):
    history.append(x)
    x -= learning_rate * gradient(x)

# Plot the function and optimization path
x_vals = np.linspace(-10, 5, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x)")
plt.scatter(history, [f(x) for x in history], color='red', label='Optimization Path', zorder=5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.axhline(0, color='gray', linewidth=0.5)
plt.title('Gradient Descent on f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 2D vector
vector = np.array([1, 0])

# Define the shear matrix along the x-axis
def shear_x(k):
    return np.array([
        [1, k],
        [0, 1]
    ])

# Define the shear matrix along the y-axis
def shear_y(k):
    return np.array([
        [1, 0],
        [k, 1]
    ])

# Define the scaling matrix
def scaling(sx, sy):
    return np.array([
        [sx, 0],
        [0, sy]
    ])

# Apply shear transformation
k_x = 1  # Shear factor for x-axis
shear_matrix_x = shear_x(k_x)
sheared_vector_x = shear_matrix_x @ vector

# Apply shear transformation (y-axis)
k_y = 1  # Shear factor for y-axis
shear_matrix_y = shear_y(k_y)
sheared_vector_y = shear_matrix_y @ vector

# Apply scaling transformation
sx = 2  # Scaling factor for x-axis
sy = 0.5  # Scaling factor for y-axis
scaling_matrix = scaling(sx, sy)
scaled_vector = scaling_matrix @ vector

# Plot the original and transformed vectors
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, sheared_vector_x[0], sheared_vector_x[1], angles='xy', scale_units='xy', scale=1, color='g', label='Sheared (X-Axis)')
plt.quiver(0, 0, sheared_vector_y[0], sheared_vector_y[1], angles='xy', scale_units='xy', scale=1, color='b', label='Sheared (Y-Axis)')
plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Scaled Vector')

# Set plot limits and details
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid()
plt.legend()
plt.title('Matrix Transformations: Shear & Scaling')
plt.show()
```
