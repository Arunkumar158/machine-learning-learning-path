```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Step 1: Create Dataset
X = np.array([1,2,3,4,5]) #Input features
y = np.array([2.2,4.1,6.0,8.1,10.1,]) #Labels (true values)
```

```python
#Step 2: Initialize Parameters
w =0.0 #Weight
b =0.0 #Bias
alpha = 0.01 #Learning rate
epochs = 100 #Number of iterations
```

```python
#Step 3: Gradient Descent
losses = [] # To store the loss at each iteration
for _ in range(epochs):
    #predict
    y_pred = w * X + b

    #Compute Loss (Mean Squared Error)
    loss = np. mean((y-y_pred)**2)
    losses.append(loss)

    #compute Gradients
    dw = (-2 / len(X)) * np.sum(X * (y - y_pred)) #Gradient wrt w
    db = (-2 / len(X)) * np.sum(y -y_pred)        #Gradient wrt b

    # update Parameters
    w = w - alpha * dw
    b = b - alpha * db

#Step 4: Plot Results
plt.plot(losses)
plt.title("Loss vs. Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()

print(f"Final weight (w): {w:.2f}")
print(f"Final bias (b): {b:.2f}")
```

```python
#visualise the line fit
plt.scatter(X, y, color='blue', label='True Data')
plt.plot(X, w * X + b, color='red', label='Fitted line')
plt.legend()
plt.show()
```
